```python
import pandas as pd
```

# merge to file, get the data from 2024.08.01 to 2024.08.31

```python
df_oasis = pd.read_csv('backend/data/1_demand_forecasting/Oasis.csv')
df_master = pd.read_csv('backend/data/1_demand_forecasting/FlightTransportsMaster.csv')
```

```python
df_oasis['Add Date'] = pd.to_datetime(df_oasis['Add Date'])
df_master['tdate'] = pd.to_datetime(df_master['tdate'])
print(df_oasis[(df_oasis['Add Date']>'2024-08-01')&(df_oasis['Add Date']<'2024-08-31')].shape,df_master[(df_master['tdate']>'2024-08-01')&(df_master['tdate']<'2024-08-31')].shape)
```

```python

df_oasis.rename(columns={'Mission Number (RC#)':'yearwithrc'},inplace=True)
```

```python

df_new = pd.merge(df_master, df_oasis, on='yearwithrc', how='inner')
```

```python
df_new = df_new[(df_new['tdate']>'2024-08-01')&(df_new['tdate']<'2024-08-31')]
```

```python
df_new.reset_index(drop=True,inplace=True)
```

```python
df_new.to_csv('backend/data/4_kpi_dashboard/merge_oasis_master_202408.csv',index=False)
```
